```python
from sklearn.datasets import make_blobs

import numpy as np
import matplotlib.pyplot as plt

def make_spirals(n_points=200, noise=0.5):
    n = np.sqrt(np.random.rand(n_points,1)) * 780 * (2*np.pi)/360
    d1x = -np.cos(n)*n + np.random.rand(n_points,1) * noise
    d1y = np.sin(n)*n + np.random.rand(n_points,1) * noise
    X = np.vstack((np.hstack((d1x,d1y)), np.hstack((-d1x,-d1y))))
    y = np.hstack((np.zeros(n_points), np.ones(n_points)))
    return X, y

X_s, y_s = make_spirals(n_points=200, noise=0.2)

plt.scatter(X_s[:,0], X_s[:,1], c=y_s, cmap=plt.cm.coolwarm)
plt.title("Dataset Spirales")
plt.show()


neural_network1(X_s.T, y_s.reshape(1,-1),hidden_layers=(128,128,128,128,32,32,32), learning_rate=0.1, n_iter=10000)
```
